```python
import os
import time
import pprint
import requests
import urllib.request
from bs4 import BeautifulSoup

url = 'https://downloads.khinsider.com/game-soundtracks/album/guilty-gear-xrd-revelator'
d_path = 'C:\\Users\\b1991\\Downloads\\'

res = requests.get(url)
#print(res.headers)
#print(res.encoding)
#print(res.content)

#soup = BeautifulSoup(res.txt, 'html.parser')で取得って書いてあるサイトもあったしそのサイトの例ではそのまま行けたけどこのGameSoundtrackのサイトではres.content使わないと取得できなかった
#https://qiita.com/Chanmoro/items/db51658b073acddea4ac
soup = BeautifulSoup(res.content, 'html.parser')
#pprint.pprint(soup.find_all('a'))

links = [url.get('href') for url in soup.find_all('a')]
#pprint.pprint(links)
#type(links)

#linksからNoneを消去
links = [e for e in links if e]

#mp3を含む要素だけ抽出
llist = [e for e in links if 'mp3' in e]

#重複を削除
llist_wodup = list(dict.fromkeys(llist))

#フルパスに変換
llist_wodup_full = ['https://downloads.khinsider.com' + e for e in llist_wodup]

i = 0
for i, e in enumerate(llist_wodup_full):
    if i > 1:
        pass
#         break

        
    mres = requests.get(e)
    msoup = BeautifulSoup(mres.content, 'html.parser')
    
    #mp3ダウンロード用本体リンク
    mlinks = [url.get('href') for url in msoup.find_all('a')]
    mlinks = [e for e in mlinks if e]
    mllist = [e for e in mlinks if 'mp3' in e]
    mllist_wodup = list(dict.fromkeys(mllist))
    m_url = mllist_wodup[0]
    print(m_url)
    
    #ファイル名作り
    #'{0:02d}'.format(i) で2桁揃え
    #[-1]で末尾にあるファイル名を取得
    #get_textでタグ文字を削除
    filename = str('{0:02d}'.format(i + 1)) + '_' + str(msoup.find_all('b')[-1].get_text() + '.mp3')
    filename = os.path.join(d_path, filename)
    print(filename)
 
    # ダウンロードする
    mem = urllib.request.urlopen(m_url).read()
 
    # ファイルへの保存
    with open(filename, mode="wb") as f:
        f.write(mem)
    
    #サイトに負荷かけすぎないようにスリープ
    time.sleep(10)
```
